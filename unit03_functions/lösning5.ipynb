# Task 1
Write a function that implements the following polynomial

p(x) = ax^3 + bx^2 + cx + d.

````python
def p1(x, a, b, c, d):
    return a*x**3 + b*x**2 + c*x + d

print(p1(2, 1, 2, 3, 4))

x = linspace(-10, 10, 100)
y = p1(x, 1, 2, 3, 4)

plot(x, y)
xlabel("x")
ylabel("p1(x)")
title("Polynom av grad 3")
grid(True)
show()
````

# Task 2
Rewrite your polynomial from the previous task such that the coefficients are represented by a list (of length 4 in this case).

````python
def p2(x, coeffs):
    return coeffs[0]*x**3 + coeffs[1]*x**2 + coeffs[2]*x + coeffs[3]
print(p2(2, [1, 2, 3, 4]))
````

# Task 3 — Packing och unpacking

Uppgiften: hur kan vi använda en *lista* med koefficienter som indata till `p1` från Task 1, trots att `p1` förväntar sig separata argument `(x, a, b, c, d)`?

**Svar:** med `*`-operatorn framför listan vid anropet — kallad **unpacking** ("packa upp").

```python
coeffs = [1, 2, 3, 4]
p1(2, *coeffs)          # ekvivalent med p1(2, 1, 2, 3, 4)
```

`*coeffs` säger till Python: *plocka ur listan och skicka in varje element som ett eget argument*. Utan `*` skulle hela listan skickas in som ett enda argument (parametern `a`), vilket ger `TypeError`.

**Motsatsen — packing:** om man istället skriver `*` i funktionens *definition*, t.ex. `def p_any(x, *coeffs):`, så *packar* Python ihop alla extra argument till en tuple. Det är användbart i Task 4 där polynomet ska kunna ha godtyckligt antal koefficienter.

| Var står `*` | Vad heter det | Vad gör det |
|--------------|---------------|-------------|
| `f(2, *lista)` (anrop) | Unpacking | Packar upp listan till separata argument |
| `def f(*args)` (definition) | Packing | Packar ihop argument till en tuple |
